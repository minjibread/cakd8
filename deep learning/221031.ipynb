```python
아침문제를 보낼 때, 매번 로그인 하는 과정이 귀찮아 로그인과 메일 발송을 한 번에 하는 프로그램을 작성하려고 한다.
아래 질문에 답해주세요.

=> RPA(Robotic Precess Automation) 업무 자동화 처리

네이버 로그인 페이지 url을 가져와 이를 웹 드라이버로 새로 호출해 주세요.

from selenium import webdriver
driver = webdriver.Chrome('./chromedriver.exe')
url = "https://nid.naver.com/nidlogin.login?mode=form&url=https%3A%2F%2Fwww.naver.com"
driver.get(url)
```

```python
아이디와 패스워드를 입력하고 엔터키까지 수행해 주세요.

i = driver.find_element(By.ID, 'id')
i.clear()
i.send_keys('id')

p = driver.find_element(By.ID, 'pw')
p.clear()
p.send_keys('pw!')

p.send_keys(Keys.ENTER)
```

```python
네이버 홈페이지에서 '메일'로 들어가도록 해 주세요.
(click event는 웹 드라이버객체 뒤에 .click() 을 해 주시면 됩니다.)

class name = 'nav'

driver.find_element(By.CLASS_NAME, 'nav').click()
```

```python
제목에 "홍길동 문제 제출입니다."라고 보내주세요.

from datetime import datetime
e = driver.find_element(By.ID, "subject")
e.clear()
now = datetime.now()
date = now.strftime('%Y-%m-%d')
date

e.send_keys(f"{date}홍길동 문제 제출입니다.") 
```
