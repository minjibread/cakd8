```python
df['구매일자'] = df['구매일자'].astype(str)
df['date'] = pd.to_datetime(df['구매일자'])
df['date']
```

# 분기별로 구분하기

```python
df = df.assign(year = df['date'].dt.year,
               month = df['date'].dt.month)
```

```python
df
```

```python
df_2014 = df.query('year == 2014')
df_2015 = df.query('year == 2015')
```

```python
print(df_2014.info(), df_2015.info())
```

```python
df_2014 = df_2014.assign(quarter = np.where(df_2014['month'] < 4, 'Q1',
                                            np.where(df_2014['month'] < 7, 'Q2',
                                                     np.where(df_2014['month'] < 10, 'Q3', 'Q4'))))
```

```python
df_2015 = df_2015.assign(quarter = np.where(df_2015['month'] < 4, 'Q5',
                                            np.where(df_2015['month'] < 7, 'Q6',
                                                     np.where(df_2015['month'] < 10, 'Q7', 'Q8'))))
```

```python
df = pd.concat([df_2014, df_2015]).sort_values(by = 'quarter')
df
```
