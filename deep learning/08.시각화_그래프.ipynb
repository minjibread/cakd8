데이터를 그래프로 표현하면 추세와 경향성이 드러나기 때문에 특징을 쉽게 이해할 수 있다.  
또, 그래프를 만드는 과정에서 새로운 패턴을 발견하기도 한다.  
특히 분석 결과를 발표할 때 그래프를 활용하면 데이터의 특징을 잘 전달할 수 있다.

# 산점도
**산점도, scatter** : 데이터를 x축과 y축에 점으로 표현한 그래프
- 변수 간 관계를 표현할 때 사용

```python
import pandas as pd
import pydataset
import seaborn as sns
```

```python
mpg = pydataset.data('mpg')
mpg
```

```python
sns.scatterplot(data = mpg, x = 'displ', y = 'hwy')
```

## 축 범위 설정하기
축은 기본적으로 최소값에서 최대값까지 모든 범위의 데이터를 표현하도록 설정되어 있다.  
일부만 표현하고 싶을 때 축 범위를 설정하자.

```python
# x축 범위 3~6으로 제한하기
sns.scatterplot(data=mpg, x='displ', y='hwy')\
   .set(xlim=(3,6))
```

```python
# x축 범위 3~6, y축 범위 10~30으로 제한하기
sns.scatterplot(data=mpg, x='displ', y='hwy')\
   .set(xlim=(3,6), ylim=(10,30))
```

## 종류별로 표식 색 바꾸기

```python
# drv별 표식 색 다르게 표현하기
sns.scatterplot(data=mpg, x='displ', y='hwy', hue='drv')
```
